## 3. Quick look at the data (2-3 real numbers)

To validate this lane selection, we analyze the raw distribution of content performance metrics inside the starter dataset to pinpoint baseline vulnerabilities.

```python
import pandas as pd
import numpy as np

# Load the starter dataset directly via raw GitHub URL to avoid path issues on Colab
csv_url = 'https://raw.githubusercontent.com/flyrank-bih/flyrank-ml-internship-starter/main/data/raw/content_refresh_anonymized.csv'
df = pd.read_csv(csv_url)

# Clean according to pipeline rules
df_clean = df[(df['impressions_90d'] > 0) & (df['content_age_days'] >= 90)].drop_duplicates(subset=['content_id'])

# Compute baseline metrics
total_pages = len(df_clean)
declining_pages = len(df_clean[df_clean['trend_direction'] == 'down'])
pct_declining = (declining_pages / total_pages) * 100

stale_pages = len(df_clean[(df_clean['content_age_days'] >= 180) & (df_clean['impressions_90d'] >= 500)])
pct_stale = (stale_pages / total_pages) * 100

print(f"--- STARTER DATASET AUDIT ---")
print(f"Total valid content pages analyzed: {total_pages}")
print(f"1. Baseline Decline Rate: {pct_declining:.2f}% of pages are in an active downward trend.")
print(f"2. Stale High-Volume Risk: {pct_stale:.2f}% of pages are older than 180 days yet sustain high impression demand.")
```
